**2.Import libraries and read the csv file**

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import pandas as pd
```

```
df=pd.read_csv('canada_per_capita_income.csv')
```

```
df.head()
```

**3. Ploting attributes in a scatter plot**

```
plt.xlabel('year', fontsize=20)
plt.ylabel('income(US$)',fontsize=20)
plt.scatter(df.year,df.income,color='red',marker='+')
```

**4. Fitting and training the model**

```
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.income)
```

```
reg.score(df[['year']],df.income)
```

```
reg.predict([[2020]])
```

```
reg.coef_
```

```
reg.intercept_
```

**5.Ploting the Linear regression plot**

```
plt.xlabel('year', fontsize=20)
plt.ylabel('income(US$)',fontsize=20)
plt.scatter(df.year,df.income,color='red',marker='+')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```
